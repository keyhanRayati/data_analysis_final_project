```python
import pandas as pd 
import warnings
warnings.filterwarnings('ignore')
```

```python
df0 = pd.read_csv("btc-1d-0.csv")
df1 = pd.read_csv("btc-1d-1.csv")
df2 = pd.read_csv("btc-1d-2.csv")
df3 = pd.read_csv("btc-1d-3.csv")
df4 = pd.read_csv("btc-1d-4.csv")

df = pd.concat([df0,df1, df2 , df3 , df4], ignore_index=True, sort=False)

```

```python
df["date"] = pd.to_datetime(df["date"])
df.dtypes
```

```python
df.sort_values(by='date' , inplace=True )
df.drop(columns = ["Unnamed: 0"] , axis = 1 , inplace=True)
df.reset_index( inplace=True)
df.drop(columns = ["index"] , axis = 1 , inplace=True)
df
```

```python
df.to_csv('btc.csv')
```

```python
df_btc = df[df['date'] >= "2015-08-07"]
df_btc.drop_duplicates(subset=['date'] , inplace=True)
df_btc.reset_index( inplace=True)
df_btc.drop(columns = ["index"] , axis = 1 , inplace=True)
df_btc

```

```python
df0 = pd.read_csv("tether-1d.csv")
df0["date"] = pd.to_datetime(df0["date"])
df0.sort_values(by='date' , inplace=True )
df0.drop(columns = ["Unnamed: 0"] , axis = 1 , inplace=True)
df0.reset_index( inplace=True)
df0.drop(columns = ["index"] , axis = 1 , inplace=True)
df0
```

```python
df_ttr = df0[df0['date'] >= "2015-08-07"]
df_ttr.reset_index( inplace=True)
df_ttr.drop(columns = ["index"] , axis = 1 , inplace=True)
df_ttr
```

```python
df_eth = pd.read_csv("eth-1d-2.csv")
df_eth["date"] = pd.to_datetime(df_eth["date"])
df_eth.sort_values(by='date' , inplace=True )
df_eth.drop(columns = ["Unnamed: 0"] , axis = 1 , inplace=True)
df_eth.reset_index( inplace=True)
df_eth.drop(columns = ["index"] , axis = 1 , inplace=True)
df_eth
```

```python
out1 = pd.merge(df_eth, df_ttr, on='date', how='outer')
data = pd.merge(out1, df_btc, on='date', how='outer')
data
```
